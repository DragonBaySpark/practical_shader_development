```python
import matplotlib.pyplot as plt

# 创建图像
fig, ax = plt.subplots()

# 假设我们要画一条从(0.2, 0.2)到(0.8, 0.8)的线，并在这条线的末尾加上箭头
start_point = (100, 100)  # 线段起点
end_point = (100, 300)    # 线段终点

# 设置x轴和y轴的范围（0到600）
plt.xlim(0, 600)
plt.ylim(0, 500)
# 设置刻度为每100一格
plt.xticks([0, 100, 200, 300, 400, 500, 600])
plt.yticks([0, 100, 200, 300, 400,500])

# 使用 annotate 函数画箭头
ax.annotate("",
            xy=end_point, xycoords='data',
            xytext=start_point, textcoords='data',
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3"))

# 设置等比例显示，避免图形变形
ax.set_aspect('equal')

plt.grid(True)


start_point = (200, 200)  # 线段起点
end_point = (200, 400)    # 线段终点
# 使用 annotate 函数画箭头
ax.annotate("",
            xy=end_point, xycoords='data',
            xytext=start_point, textcoords='data',
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3",ec='green'))


start_point = (0, 0)  # 线段起点
end_point = (0, 200)    # 线段终点
# 使用 annotate 函数画箭头
ax.annotate("",
            xy=end_point, xycoords='data',
            xytext=start_point, textcoords='data',
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3",ec='red'))

# 显示图像
plt.show()
```

```python
import matplotlib.pyplot as plt 

def draw_line(ax,start_point,end_point,color,text=None):
    ax.annotate("",
            xy=end_point, xycoords='data',
            xytext=start_point, textcoords='data',
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3",ec=color))
    if text:
        # 计算中点坐标
        mid_x = (start_point[0] + end_point[0]) / 2-0.1
        mid_y = (start_point[1] + end_point[1]) / 2
        # 添加文字，设置偏移量避免遮挡线段
        ax.text(mid_x, mid_y + 0.2, text, ha='center', va='bottom')

fig,ax=plt.subplots()
ax.set_aspect('equal')
plt.grid()
ax.set(xlim=(0,5),ylim=(0,5))

o=(0,0)
a=(4,1)
b=(4,4)


draw_line(ax,o,a,'r','A')
draw_line(ax,o,b,'g','C')
draw_line(ax,a,b,'b','B')





plt.show()
```
